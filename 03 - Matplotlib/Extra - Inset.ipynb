```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1, 0.1, 0.35, 0.35])
ax3 = fig.add_axes([0.6, 0.6, 0.35, 0.35])

ax1.plot(x, y, '-o')
ax2.plot(x[10:15], y[10:15], '-o')
ax3.plot(x[35:40], y[35:40], '-o')
```
